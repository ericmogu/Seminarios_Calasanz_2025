# Ejemplo de aplicación POO

Un profesor debe calcular el promedio de la nota de las evaluaciones de sus estudiantes para subirla a la plataforma de notas finales.
Sin embargo, el profesor acordó con sus estudiantes que los ayudará eliminando la peor de las 5 notas antes de calcular el promedio que finalmente reportará. Adicionalmente, el profesor se ha dado cuenta qué las notas registradas en su planilla se encuentran en una escala de números enteros de 0 a 100 pero la plataforma está diseñada para recibir el promedio únicamente en la escala estándar de la universidad: de 0 a 5, redondeado a dos decimales.

```python
from datetime import datetime

class Nota:
    #Constructor(es) 
    ######################
    def __init__(self, nota5: float = None, nota100: int = None, descripcion: str = None):
        if not(nota100 == None):
            self.__nota100 = nota100
            self.convertirNota_100_5()
        else:
            self.__nota5 = nota5
            self.convertirNota_5_100()

        self.__descripcion = descripcion
        self.__fechaRegistro = datetime.now()
        self.__notaCualitativa = None
        
        self.convertirNota_100_Cualitativa()
    
    #Métodos (comportamiento del objeto)
    ################################### 
    def convertirNota_5_100(self):
        self.__nota100 = round((self.__nota5 * 100)/5)
    
    def convertirNota_100_5(self):
        self.__nota5 = round((self.__nota100 * 5)/100,1)
    
    def convertirNota_100_Cualitativa(self):
        if self.__nota100 > 60:
            self.__notaCualitativa = 'Aprobado'
        else:
            self.__notaCualitativa = 'Reprobado'
    
    #Getters
    ########
    def getNota5(self) -> float:
        return self.__nota5
    
    def getNota100(self) -> float:
        return self.__nota100
    
    def getNotaCualitativa(self) -> float:
        return self.__notaCualitativa
    
    def getFechaRegistro(self) -> float:
        return str(self.__fechaRegistro)
    
    def getDescripcion(self) -> float:
        return self.__descripcion
    
    #Setters
    ########
    def setNota5(self, nota5: float):
        self.__nota5 = round(nota5,1)
        self.convertirNota_5_100()
        self.convertirNota_100_Cualitativa()
    
    def setNota100(self, nota100: int):
        self.__nota100 = round(nota100)
        self.convertirNota_100_5()
        self.convertirNota_100_Cualitativa()
    
    def setDescripcion(self, descripcion: str):
        self.__descripcion = descripcion

    # Devolver la información del objeto
    ################################################
    def __str__(self) -> str:
        mensaje = "------------------------\n"
        mensaje += f"Descripción: {self.getDescripcion()}\n"
        mensaje += f"Fecha Registro: {self.getFechaRegistro()}\n"
        mensaje += f"Nota Cualitativa: {self.getNotaCualitativa()}\n"
        mensaje += f"Nota Escala 100: {self.getNota100()}\n"
        mensaje += f"Nota Escala 5: {self.getNota5()}"
        return mensaje
```

```python
from functools import reduce

class Materia:    
    #Constructor(es) 
    ######################
    def __init__(self, codigo: str, nombre: str, semestre: int, creditos: int, *notas):
        self.__codigo = codigo
        self.__nombre = nombre
        self.__semestre = semestre
        self.__creditos = creditos
        self.__notas = list(filter(lambda x: x != None, notas))

    #Métodos (comportamiento del objeto)
    ################################### 
    def adicionarNota(self, nota: Nota):
        if nota != None:
            self.__notas.append(nota)
    
    #Getters
    ########
    def getPromedio5(self):
        notas = list(map(lambda x: x.getNota5(), self.__notas))
        return round(sum(notas)/len(notas) , 2)
    
    def getPromedioAjustado5(self):
        notas = list(map(lambda x: x.getNota5(), self.__notas))
        return round((sum(notas) - min(notas)) / (len(notas)-1), 2)
    
    def getPromedio100(self):
        notas = list(map(lambda x: x.getNota100(), self.__notas))
        return round(sum(notas)/len(notas) , 2)
    
    def getPromedioAjustado100(self):
        notas = list(map(lambda x: x.getNota100(), self.__notas))
        return round((sum(notas) - min(notas)) / (len(notas)-1), 2)
    
    #Setters
    ########   

    # Devolver la información del objeto
    ################################################
    def __str__(self) -> str:
        mensaje = "------------------------\n"
        mensaje += f"Código: {self.__codigo}\n"
        mensaje += f"Curso: {self.__nombre}\n"
        mensaje += f"Semestre: {self.__semestre}\n"
        mensaje += f"Créditos: {self.__creditos}\n"
        mensaje += f"# Notas: {len(self.__notas)}\n"
        mensaje += f"Promedio5: {self.getPromedio5()}\n"
        mensaje += f"Promedio100: {self.getPromedio100()}\n"
        mensaje += f"Promedio5Ajustado: {self.getPromedioAjustado5()}\n"
        mensaje += f"Promedio100Ajustado: {self.getPromedioAjustado100()}"
        return mensaje

    def mostrarNotasMateria(self):
        for nota in self.__notas:
            print(nota)    
```

```python
#Instanciar: crear los objetos tipo nota
objetoNota1 = Nota(nota100=40,descripcion='Primer Parcial')
objetoNota2 = Nota(nota100=50,descripcion='Segundo Parcial')
objetoNota3 = Nota(nota100=39,descripcion='Tercer Parcial')

#Pedir al objeto que se muestre en pantalla (consola)
print("***Notas")
print(objetoNota1, objetoNota2, objetoNota3, sep="\n")
```

```python
#Instanciar: crear los objetos tipo nota
objetoMateria = Materia("10001", "Estructura de Lenguajes", 1, 3, objetoNota1, objetoNota2, objetoNota3)

#Solicitar a la materia que se muestre en pantalla
print("***Materia (Respuesta al requerimiento)")
print(objetoMateria)
```

```python
objetoMateria.adicionarNota(Nota(nota100=76,descripcion='Cuarto Parcial'))
objetoMateria.adicionarNota(Nota(nota100=96,descripcion='Quinto Parcial'))
print(objetoMateria)
```

```python
objetoMateria.mostrarNotasMateria()
```
